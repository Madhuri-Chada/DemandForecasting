```python
outliers=[]
def detect_outliers(forecast_value):
  
    threshold = 1
    mean = np.mean(forecast_value)
    std = np.std(forecast_value)
    
    
    for i in forecast_value:
        z_score = (i - mean)/std
        if np.abs(z_score) > threshold:
            outliers.append(i)
    return outliers
```

```python
def smape(actual_value, forecast_value):
    denominator = (np.abs(actual_value) + np.abs(forecast_value))
    diff = np.abs(actual_value - forecast_value) / denominator
    diff[denominator == 0] = 0.0
    return 200 * np.mean(diff)
```
