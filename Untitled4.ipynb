```python
def smape(actual_value, forecast_value):
    denominator = (np.abs(actual_value) + np.abs(forecast_value))
    diff = np.abs(actual_value - forecast_value) / denominator
    diff[denominator == 0] = 0.0
    return 100 * np.mean(diff)
```

```python
smape_value=smape(actual_value, forecast_value)
smape_value
```

```python
 outliers=[]
def detect_outliers(df_stationery_convert):
  
    threshold = 3
    mean = np.mean(df_stationery_convert)
    std = np.std(df_stationery_convert)
    
    
    for i in df_stationery_convert:
        z_score = (i - mean)/std
        if np.abs(z_score) > threshold:
            outliers.append(i)
    return outliers
```

```python
df_list = df_stationery_convert
outlier_pt = detect_outliers(df_stationery_convert)
outlier_pt
```
